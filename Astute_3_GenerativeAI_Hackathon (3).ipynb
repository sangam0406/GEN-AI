```python
#importing libraries
import whisper
import os
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
from typing_extensions import Concatenate
from langchain.chains.question_answering import load_qa_chain
from langchain.llms import OpenAI
```

<div style="display:flex ">
    <div style="margin: auto auto auto 0"><h4>Implemented using English Video</h2>  </div>
</div>

```python
#defining paths
video_file_path = 'meet.mp4'
mp3_file_path = 'meet.mp3'

# Convert video to MP3
command_to_mp3 = f"ffmpeg -i {video_file_path} {mp3_file_path}"
os.system(command_to_mp3)
```

```python
# using openai whisper to converting video to transcription
model=whisper.load_model("small")
result=model.transcribe(mp3_file_path)
```

```python
#showcasing trascription
raw_text=result["text"]
raw_text
```

```python
# dividing transcription into chunks for proper embedding
text_splitter=CharacterTextSplitter(
    separator="\n",
    chunk_size = 100,
    chunk_overlap = 20,
)
texts=text_splitter.split_text(raw_text)
```

```python
#using HuggingFace embedding for embedding text
embeddings= HuggingFaceEmbeddings()
```

```python
#using FAISS vectorstore to save embedded values
document_search=FAISS.from_texts(texts,embeddings)
document_search
```

```python
#storing openai-api key
os.environ["OPENAI_API_KEY"]="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"
```

```python
# establishing question answering chain for connecting LLM with Vectorstore
chain=load_qa_chain(OpenAI(),chain_type="stuff")
```

```python
#providing query
query="Provide important points of this meetings and what are the task that needs to be done"
#using similarity search for finding similar embedded values relevant to user queries
docs=document_search.similarity_search(query)
#running chain
answer=chain.run(input_documents=docs, question=query)
#displaying answers
print(answer)
```
